#### Create 3 lions


```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

image = Image.open("Belgium_lion.png").convert('RGB') 
image = image.resize((250, 250))
image_array = np.array(image)

black_mask = (image_array[:, :, 0] < 100) & (image_array[:, :, 1] < 100) & (image_array[:, :, 2] < 100)

red_lion = image_array.copy()
blue_lion = image_array.copy()
green_lion = image_array.copy()

red_lion[black_mask] = [255, 0, 0]
blue_lion[black_mask] = [0, 0, 255]
green_lion[black_mask] = [0, 255, 0]

fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].imshow(red_lion)
axes[0].set_title("Red Lion")
axes[0].axis("off")

axes[1].imshow(blue_lion)
axes[1].set_title("Blue Lion")
axes[1].axis("off")

axes[2].imshow(green_lion)
axes[2].set_title("Green Lion")
axes[2].axis("off")

plt.show()
```

#### Gradient functie

```python
def gradient(image):
    """Applies an orange gradient over the entire image."""
    
    height, width, channels = image.shape
    gradient = image.copy()

    orange_gradient = np.linspace(0, 255, width).astype(np.uint8)
    orange_gradient = np.tile(orange_gradient, (height, 1))

    gradient[:, :, 0] = np.maximum(gradient[:, :, 0], orange_gradient)
    gradient[:, :, 1] = np.maximum(gradient[:, :, 1], (orange_gradient * 0.6))  

    return gradient
```

```python
gradient_image = gradient(red_lion)

plt.imshow(gradient_image)
plt.axis("off")
plt.show()
```

### Sine wave functie

```python
def sine_distortion(image, amplitude=20, frequency=0.05):
    
    height, width, channels = image.shape

    # Generate coordinate grids
    x_coords = np.tile(np.arange(width), (height, 1))  # Width x Height array
    y_coords = np.tile(np.arange(height), (width, 1)).T  # Transposed to match image shape

    transformed_x = x_coords + amplitude * np.sin(2 * np.pi * frequency * y_coords)
    transformed_x = np.clip(transformed_x.astype(int), 0, width - 1)
    
    distorted_image = np.zeros_like(image)

    for channel in range(channels):
        distorted_image[:, :, channel] = image[y_coords, transformed_x, channel]

    return distorted_image.astype(np.uint8)
```

```python
distorted_image = sine_distortion(red_lion)

plt.imshow(distorted_image)
plt.axis("off")
plt.show()

```
